```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

```python
#s=Service(r"X:/CampusX/projects/laptopdataset_eda/chrome-win64/chrome.exe")
driver=webdriver.Chrome()
driver.get("https://www.smartprix.com/laptops")
time.sleep(3)
old_height = driver.execute_script('return document.body.scrollHeight')
while True:

    driver.find_element(by=By.XPATH, value='//*[@id="app"]/main/div[1]/div[3]/div[3]').click()
    time.sleep(2)

    new_height = driver.execute_script('return document.body.scrollHeight')

    if new_height == old_height:
        break

    old_height = new_height

```

```python
html = driver.page_source

with open('smartprix.html','w',encoding='utf-8') as f:
    f.write(html)
```

```python
with open('smartprix.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
containers=soup.find_all("div","sm-product has-tag has-features has-actions")
```

```python
name=[]
price=[]
ratings=[]
num_rating_reviews=[]
general=[]
display=[]
performance=[]
connectivity=[]
battery=[]
others=[]
```

```python
for i in containers[689:]:
    anchor_tags=i.find("a")
    half_link=anchor_tags.get("href")
    link= "https://www.smartprix.com{}".format(half_link)
    driver=webdriver.Chrome()
    driver.get(link)
    time.sleep(3)
    extractor()
    driver.close()
    
    
        
    
    
    
```

```python
def extractor():
    try:
        name.append(driver.find_element(by=By.XPATH, value='//*[@id="app"]/main/div[1]/div[2]/div[1]/div[1]/h1').text)
    except:
        name.append(np.nan)
    try:
        price.append(driver.find_element(by=By.XPATH, value='//*[@id="app"]/main/div[1]/div[2]/div[1]/div[1]/div/div[2]/strong').text)
    except:
        price.append(np.nan)
    try:
        ratings.append(driver.find_element(by=By.XPATH, value='//*[@id="app"]/main/div[1]/div[2]/div[1]/div[1]/div/div[1]/span[2]').text)
    except:
        ratings.append(np.nan)
    try:
        num_rating_reviews.append(driver.find_element(by=By.XPATH, value='//*[@id="app"]/main/div[1]/div[2]/div[1]/div[1]/div/div[1]/span[1]').get_attribute('style'))
    except:
        num_rating_reviews.append(np.nan)
    html = driver.page_source
    soup=BeautifulSoup(html,"lxml")
    specs=soup.find("div",{"class":"sm-quick-specs"}).find_all("ul")
    list=[]
    if len(specs)==5:
        for j in specs[0].find_all("li"):
            list.append(j.text)
        general.append(list)
        list=[]
        for j in specs[1].find_all("li"):
            list.append(j.text)
        display.append(list)
        list=[]
        for j in specs[2].find_all("li"):
            list.append(j.text)
        performance.append(list)
        list=[]
        for j in specs[3].find_all("li"):
            list.append(j.text)
        connectivity.append(list)
        list=[]
        for j in specs[4].find_all("li"):
            list.append(j.text)
        battery.append(list)
        others.append(np.nan)
    elif len(specs)>5:
        for j in specs[0].find_all("li"):
            list.append(j.text)
        general.append(list)
        list=[]
        for j in specs[1].find_all("li"):
            list.append(j.text)
        display.append(list)
        list=[]
        for j in specs[2].find_all("li"):
            list.append(j.text)
        performance.append(list)
        list=[]
        for j in specs[3].find_all("li"):
            list.append(j.text)
        connectivity.append(list)
        list=[]
        for j in specs[4].find_all("li"):
            list.append(j.text)
        battery.append(list)
        list=[]
        for j in specs[5].find_all("li"):
            list.append(j.text)
        others.append(list)
        
    else:
        for j in specs[0].find_all("li"):
            list.append(j.text)
        general.append(list)
        list=[]
        for j in specs[1].find_all("li"):
            list.append(j.text)
        display.append(list)
        list=[]
        for j in specs[2].find_all("li"):
            list.append(j.text)
        performance.append(list)
        list=[]
        for j in specs[3].find_all("li"):
            list.append(j.text)
        connectivity.append(list)
        battery.append(np.nan)
        others.append(np.nan)
        
    

    
    
```

```python
df=pd.DataFrame({
    "name":name,
    "price":price,
    "ratings":ratings,
    "num_rating_reviews":num_rating_reviews,
    "general":general,
    "display":display,
    "performance":performance,
    "connectivity":connectivity,
    "battery":battery,
    
})
```

```python
df.to_csv("smartprix_laptop.csv")
```
